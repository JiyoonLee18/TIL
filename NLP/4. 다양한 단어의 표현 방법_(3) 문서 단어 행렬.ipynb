## 문서 단어 행렬(Document-Term Matrix, DTM)
- https://wikidocs.net/24559
### 문서 단어 행렬(Document-Term Matrix, DTM)의 표기법
- 문서 단어 행렬(Document-Term Matrix, DTM): 다수의 문서에서 등장하는 각 단어들의 빈도를 행렬로 표현한 것
- 각 문서에 대한 BoW를 하나의 행렬로 만든 것
- BoW와 다른 표현 방법이 아니라 BoW 표현을 다수의 문서에 대해서 행렬로 표현하고 부르는 용어

- ex. 띄어쓰기 단위 토큰화, 문서 단어 행렬로 표현
    - 문서1 : 먹고 싶은 사과
    - 문서2 : 먹고 싶은 바나나
    - 문서3 : 길고 노란 바나나 바나나
    - 문서4 : 저는 과일이 좋아요
    
    - 각 문서에서 등장한 단어의 빈도를 행렬의 값으로 표기
    - 서로 비교할 수 있도록 수치화할 수 있다는 점에서 의의

### 문서 단어 행렬(Document-Term Matrix)의 한계
- DTM은 매우 간단하고 구현하기도 쉽지만, 본질적으로 가지는 몇 가지 한계 존재

#### 희소 표현(Sparse representation)
- 원-핫 벡터는 단어 집합의 크기가 벡터의 차원이 되고 대부분의 값이 0이 되는 벡터
    - 원-핫 벡터는 공간적 낭비와 계산 리소스를 증가시킬 수 있다는 점에서 단점
- DTM도 마찬가지, 각 문서 벡터의 차원은 전체 단어 집합의 크기를 가짐. 방대한 데이터일수록 아주 큰 차원 가짐.
    - 또한 많은 문서 벡터가 대부분의 값이 0을 가질 수도 있음
- 희소 벡터(sparse vector) 또는 희소 행렬(sparse matrix): 원-핫 벡터나 DTM과 같은 대부분의 값이 0인 표현
    - 희소 벡터는 많은 양의 저장 공간과 높은 계산 복잡도를 요구
    - 이로 인해 처리를 통해 단어 집합의 크기를 줄이는 일은 BoW 표현을 사용하는 모델에서 중요
    - 구두점, 빈도수가 낮은 단어, 불용어를 제거하고, 어간이나 표제어 추출을 통해 단어를 정규화하여 단어 집합의 크기를 줄이기 가능

#### 단순 빈도 수 기반 접근
- 여러 문서에 등장하는 모든 단어에 대해서 빈도 표기는 한계점 있음
    - ex. 영어 DTM의 경우, 불용어인 the는 어떤 문서이든 자주 등장, the의 빈도수가 높다고 해서 유사한 문서라고 판단 불가
- 각 문서에는 중요한 단어와 불필요한 단어들이 혼재
    - 불용어(stopwords)와 같은 단어들은 빈도수가 높더라도 자연어 처리에 있어 의미를 갖지 못하는 단어 존재
    - TF-IDF: DTM에 불용어와 중요한 단어에 대해서 가중치를 줄 수 있는 방법
