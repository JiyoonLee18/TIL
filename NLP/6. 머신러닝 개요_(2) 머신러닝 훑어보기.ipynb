### 머신러닝 모델의 평가

- 데이터를 훈련용, 검증용, 테스트용 이렇게 세 가지로 분리하는 것이 일반적
    - 훈련 데이터: 머신 러닝 모델을 학습하는 용도
    - 테스트 데이터: 학습한 머신 러닝 모델의 성능을 평가하기 위한 용도
    - 검증용 데이터: 모델의 성능을 평가하기 위한 용도가 아니라 모델의 성능을 조정하기 위한 용도, 훈련 데이터에 과적합(overfitting) 이 되고 있는지 판단하거나 하이퍼파라미터의 조정을 위한 용도
        - 하이퍼파라미터(초매개변수) : 모델의 성능에 영향을 주는 사람이 값을 지정하는 변수
            - 하이퍼파라미터는 보통 사용자가 직접 정해줄 수 있는 변수라는 점
            - ex. 학습률 learning rate, 뉴런의 수나 층의 수
        - 매개변수 : 가중치와 편향. 학습을 하는 동안 값이 계속해서 변하는 수
            - ex. 가중치와 편향
            - 사용자가 결정해주는 값이 아니라 모델이 학습하는 과정에서 얻어지는 값
        - 훈련용 데이터로 훈련을 모두 시킨 모델은 검증용 데이터를 사용하여 정확도를 검증하며 튜닝함
        - 검증용 데이터에 대해서 높은 정확도를 얻도록 파라미터의 값을 바꿔보는 것, 튜닝하는 과정에서 모델은 검증용 데이터의 정확도를 높이는 방향으로 점차적으로 수정됨

### 분류(Classification)와 회귀(Regression)
- 머신러닝의 많은 문제: 분류 또는 회귀 문제
    - 선형 회귀는 대표적인 회귀 문제
    - 로지스틱 회귀는 (이름은 회귀이지만) 대표적인 분류 문제
    - 이진 분류(Binary Classification)과 다중 클래스 분류(Multi-Class Classification)
        - 엄밀히는 다중 레이블 분류(Multi-lable Classification)라는 또 다른 문제가 존재하지만 다루지 x
- 1) 이진 분류 문제(Binary Classification)
    - 이진 분류는 주어진 입력에 대해서 두 개의 선택지 중 하나의 답을 선택
    - ex. 성적표 합격, 불합격
    - ex. 정상메일, 스팸메일
- 2) 다중 클래스 분류(Multi-class Classification)
    - 주어진 입력에 대해서 세 개 이상의 선택지 중에서 답을 선택해야 하는 경우
    - ex. 책을  과학, 영어, IT, 학습지, 만화로 분류
- 3) 회귀 문제(Regression)
    - 어떠한 연속적인 값의 범위 내에서 예측값이 나오는 경우
    - ex. 역과의 거리, 인구 밀도, 방의 개수 등을 입력하면 부동산 가격을 예측
    - ex. 시계열 데이터(Time Series Data)를 이용한 주가 예측, 생산량 예측

### 지도 학습과 비지도 학습
- 머신 러닝은 크게 지도 학습, 비지도 학습, 강화 학습으로 나눔
    - (강화학습의 경우 책 범위 x)
- 1) 지도 학습(Supervised Learning)
    - 지도 학습: 레이블(Label)이라는 정답과 함께 학습하는 것
    - 자연어 처리는 대부분 지도 학습에 속함
    - 레이블은 'y', '실제값' 등으로 불리기도 함
- 2) 비지도 학습(Unsupervised Learning)
    - 데이터에 별도의 레이블이 없이 학습하는 것
    - 텍스트 처리 분야의 토픽 모델링 알고리즘인 LSA나 LDA는 비지도학습
- 3) 자기지도 학습(Self-Supervised Learning, SSL)
    - 레이블이 없는 데이터가 주어지는 경우, 모델이 학습을 위해서 스스로 데이터로부터 레이블을 만들어서 학습
    - ex. Word2Vec과 같은 워드 임베딩 알고리즘
    - ex. BERT와 같은 언어 모델의 학습 방법

### 샘플(Sample)과 특성(Feature)
- 대부분의 머신러닝: 1개 이상의 독립변수 x로 종속 변수 y를 예측
    - 특히, 인공 신경망은 독립 변수, 종속 변수, 가중치, 편향 등을 행렬 연산을 통해 연산하는 경우가 다수
    - 훈련 데이터를 행렬로 표현하는 경우
- 독립변수 x의 행렬을 x라고 할 때, 독립변수의 개수가 n이고 데이터의 개수가 m인 행렬x는 다음 이미지와 같음


- 샘플: 하나의 데이터, 행렬 관점에서는 하나의 행을 일컬음
- 특성: 종속변수 y를 예측하기 위한 각각의 독립변수 x, 행렬 관점에서는 각 열에 해당

### 혼동 행렬(Confusion Matrix)
- 정확도(Accuracy): 맞춘 문제수를 전체 문제수로 나눈 값
- 혼동 행렬: 맞춘 결과와 틀린 결과에 대한 세부적인 내용을 알려주는 행렬

    - True Positive(TP) : 실제 True인 정답을 True라고 예측 (정답)
    - False Positive(FP) : 실제 False인 정답을 True라고 예측 (오답)
    - False Negative(FN) : 실제 True인 정답을 False라고 예측 (오답)
    - True Negative(TN) : 실제 False인 정답을 False라고 예측 (정답)
- 정밀도(Precision): 모델이 True라고 분류한 것 중에서 실제 True인 것의 비율
    - 정밀도=TP/(TP+FP)
- 재현율(Recall):  실제 True인 것 중에서 모델이 True라고 예측한 것의 비율
    - 재현율=TP/(TP+FN)
- 정확도(Accuracy): 전체 예측한 데이터 중에서 정답을 맞춘 것에 대한 비율
    - 정확도=TP+TN/(TP+FN+FP+TN)
    - Accuracy로 성능을 예측하는 것이 적절하지 않은 때가 존재
        - ex. 200일 동안 총 6일만 비가 왔는데, 모델이 200일 내내 맑았다가 예측, 이 경우도 정확도는 97%
    - 실질적으로 더 중요한 경우에 대한 데이터가 전체 데이터에서 너무 적은 비율을 차지하는경우, F1-SCORE 사용
- F1-SCORE
    - 2*정밀도*재현율/정밀도+재현율=TP/(TP+(FN+FP)/2)

### 과적합(Overfitting)과 과소 적합(Underfitting)
- 과적합(Overfitting): 훈련 데이터를 과하게 학습한 경우
    - 기계가 훈련 데이터에 대해서만 과하게 학습하면 성능 측정을 위한 데이터인 테스트 데이터나 실제 서비스에서는 정확도가 좋지 않은 현상이 발생
    - 훈련 데이터에 대해서는 오차가 낮지만, 테스트 데이터에 대해서는 오차가 커지는 경우
    
    - 위 그래프의 경우 30 에포크로 학습하여 과적합 발생 모습
    - 에포크 3~4에서 테스트 데이터에 대한 정확도가 가장 높고, 에포크가 그 이상 넘어가면 과적합이 발생
- 과소적합(Underfitting): 테스트 데이터의 성능이 올라갈 여지가 있음에도 훈련을 덜 한 상태
    - 과소 적합은 훈련 자체가 부족한 상태이므로 훈련 횟수인 에포크가 지나치게 적으면 발생
- 적합(fitting): 머신 러닝에서 학습 또는 훈련이라고 하는 과정 (과+적합, 과소+적합)
