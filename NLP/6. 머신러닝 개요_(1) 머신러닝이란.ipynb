# 머신 러닝(Machine Learning) 개요
- 머신 러닝은 영상 처리, 번역기, 음성 인식, 스팸 메일 탐지 등 굉장히 다양한 분야에서 응용되고 있음
- 해당 챕터: 머신 러닝의 개념과 선형 회귀, 로지스틱 회귀, 소프트맥스 회귀와 같은 기본적인 모델 다룸

## 머신 러닝이란(What is Machine Learning?)
- https://wikidocs.net/21679
- 딥 러닝을 포함하고 있는 개념인 머신 러닝의 개념

### 머신러닝 방식

- 위 이미지: 머신러닝이 아닌 기존의 프로그램의 방식의 경우
- 아래 이미지: 머신러닝의 형식
- 머신 러닝은 데이터가 주어지면, 기계가 스스로 데이터로부터 규칙성을 찾는 것에 집중
- 주어진 데이터로부터 규칙성을 찾는 과정을 우리는 훈련(training) 또는 학습(learning)라고 함
- 일단 규칙성을 발견하고나면, 그 후에 들어오는 새로운 데이터에 대해서 발견한 규칙성을 기준으로 정답을 찾아내는 방식
