# 언어모델
- https://wikidocs.net/21695
- 언어 모델: 단어 시퀀스(문장)에 확률을 할당하는 모델
- 전통적인 언어 모델(Statistical Languagel Model, SLM)
    - 통계에 기반한 언어 모델은 한계로 인해 사용 용도가 감소함
    - 하지만 통계 기반 방법론에 대한 이해는 언어 모델에 대한 이해도를 향상시켜주기에 도움이 됨

## 언어 모델(Language Model)이란?
- 언어 모델을 만드는 방법은 크게는 통계를 이용한 방법과 인공 신경망을 이용한 방법으로 구분 가능함
- 인공 신경망을 이용한 방법이 더 좋은 성능을 보여줌
    - GPT나 BERT 또한 인공 신경망 언어 모델의 개념을 사용함

### 언어 모델(Language Model)
- 언어 모델은 단어 시퀀스에 확률을 할당(assign) 하는 일을 하는 모델
    - 가장 자연스러운 단어 시퀀스를 찾아내는 모델
    -  단어 시퀀스에 확률을 할당하게 하기 위해서 가장 보편적으로 사용되는 방법: 이전 단어들이 주어졌을 때 다음 단어를 예측
- 언어 모델링(Language Modeling)
    - 주어진 단어들로부터 아직 모르는 단어를 예측하는 작업

### 단어 시퀀스의 확률 할당
- 언어 모델은 위와 같이 확률을 통해 보다 적절한 문장을 판단합니다.
- 기계 번역(Machine Translation)
    - ex. P(나는 버스를 탔다) > P(나는 버스를 태운다)
- 오타 교정(Spell Correction)
    - ex. P(달려갔다) > P(잘려갔다)
- 음성 인식(Speech Recognition)
    - ex. P(나는 메롱을 먹는다) < P(나는 메론을 먹는다)
