## 선형 회귀(Linear Regression)
- https://wikidocs.net/21670

### 선형 회귀(Linear Regression)
- 다른 변수의 값을 변하게하는 변수를 x, 변수 x에 의해서 값이 종속적으로 변하는 변수를 y
- 선형 회귀는 한 개 이상의 독립 변수 x와 y의 선형 관계를 모델링

#### 단순 선형 회귀 분석(Simple Linear Regression Analysis)
- y=wx+b
- 독립변수 x와 곱해지는 값 w를 머신러닝에서는 가중치 weight, 별도로 더해지는 값 b를 편향 bias 라고 함
- 직선의 방정식에서는 각각 직선의 기울기와 절편을 의미함

#### 다중 선형 회귀 분석(Multiple Linear Regression Analysis)
- y=w1x1+w2x2+...+wnxn+b
- y는 여전히 1개이지만 이제 x는 1개가 아니라 여러 개

### 가설(Hypothesis) 세우기
- ex. 단순 선형 회귀일때,
- x와 y의 관계를 유추하기 위해 수학적으로 식을 세움
- 이러한 식을 가설 Hypothesis 이라고 함
- H(x)=wx+b


- w와 b의 값에 따라서 직선은 다양하게 그려짐
- 선형 회귀에서 해야하는 일 = 적절한 w와 b를 찾는 일 (데이터를 잘 표현하는)

### 비용 함수(Cost function) : 평균 제곱 오차(MSE)
- 적절한 w와 b를 찾기 위해서, 머신러닝은 실제값과 가설로부터 얻은 예측값의 오차를 계산하는 식을 세우고, 이 식의 값을 최소화하는 최적의 w와 b를 찾아냄
- 목적 함수(Objective function), 비용 함수(Cost function), 손실 함수(Loss function): 실제값과 예측값에 대한 오차에 대한 식
    - 함수의 값을 최소화하거나, 최대화하거나 하는 목적을 가진 함수
    - 값을 최소화
    - 단순히 실제값과 예측값에 대한 오차를 표현하면 되는 것이 아니라, 예측값의 오차를 줄이는 일에 최적화 된 식
- 평균 제곱 오차(Mean Squared Error, MSE): 회귀문제에서 주로 사용


### 옵티마이저(Optimizer) : 경사하강법(Gradient Descent)
- 머신 러닝, 딥 러닝의 학습은 결국 비용 함수를 최소화하는 매개 변수인 w와 b을 찾기 위한 작업을 수행
    - 이 때 사용되는 알고리즘을 옵티마이저(Optimizer) 또는 최적화 알고리즘이라고 부름
    - 이 옵티마이저를 통해 적절한 w와 b를 찾아내는 과정을 머신 러닝에서 훈련(training) 또는 학습(learning)이라고 부름
- 경사 하강법(Gradient Descent): 가장 기본적인 옵티마이저 중 하나
    - 설명의 편의를 위해 편향 b없이 단순히 가중치w만을 사용한 y=wx을 가지고 경사 하강법 수행한다면,
    
    - 기울기w가 무한대로 커지면 커질 수록 cost의 값 또한 무한대로 커지고, 반대로 기울기 w가 무한대로 작아져도 cost의 값은 무한대로 커짐
    - cost가 가장 작을때는 블록한 부분의 맨 아래부분
    
    - cost가 최소화가 되는 지점은 접선의 기울기가 0이 되는 지점(미분 활용)
    - 비용을 최소화하기 위해 w를 구하기 위해 w를 업데이트 하는 식
    
    
    - a는 학습률 learning rate를 의미함
        - 학습률a는 w의 값을 변경할 때, 얼마나 크게 변경할지를 결정
        - 0과 1 사이의 값을 가짐
        - 무조건 크다고 빠르게 찾는 것은 아님
